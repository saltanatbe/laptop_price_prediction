### dividing the information about the screen resolution

```python
#create touchscreena dn screenresolution columns
df['Touchscreen'] = df['ScreenResolution'].apply(lambda element : 1 if 'Touchscreen' in element else 0)
df['IPS'] = df['ScreenResolution'].apply(lambda element : 1 if 'IPS' in element else 0)

df.head()
```

```python
sn.barplot(x=df['Touchscreen'], y=df.Price)
```

```python
sn.barplot(x=df['IPS'], y=df.Price)
```

### extract the size resolution of the screen (x resolution and y resolution)

```python
resolution = df.ScreenResolution.str.split('x', n=1, expand=True)
df['X_resolution'] = resolution[0].str.replace(',', '').str.findall(r'(\d+\.?\d+)').apply(lambda element : element[0]).astype('int64')
df['Y_resolution'] = resolution[1].astype('int64')
df
```

```python
sn.heatmap(df.corr())
```

```python
df.corr()['Price']
```

from the correlation it can be seen that both x resolution and y resolution are both dependent on each other and are affecting the price the same. There shouldn't be two features affecting the same way. So we perform future reduction and create new column pixels per inch feature

```python
df['ppi'] = (((df['X_resolution']**2+df['Y_resolution']**2))**0.5/df['Inches']).astype('float')
df.info()
```

```python
df.corr()['Price']
```

pixels per inch (ppi) now is the combination of three features, therefore they can be eliminated

```python
df.drop(columns=['X_resolution', 'Y_resolution', 'Inches'], inplace = True)
df
```

```python
memory_df = df['Memory'].astype(str).replace('\.0', '', regex=True)
memory_df = memory_df.str.replace('GB', '')
memory_df = memory_df.str.replace('TB', '000')
memory_df = memory_df.str.split('+', expand=True)
memory_df[1] = memory_df[1].fillna("0")
memory_df
```

```python
def whichMemory(memory):
    memory_df["0 " + memory] = memory_df[0].apply(lambda x : x if memory in x else 0)
    memory_df["1 " + memory] = memory_df[1].apply(lambda x : x if memory in x else 0)
    memory_df["0 " + memory] = memory_df["0 " + memory].replace('\D', '', regex=True).astype(int)
    memory_df["1 " + memory] = memory_df["1 " + memory].replace('\D', '', regex=True).astype(int)
    memory_df[memory] = memory_df["0 " + memory] +  memory_df["1 " + memory]
    df[memory] = memory_df[memory]
        
for i in ['SSD', 'HDD', 'Hybrid', 'Flash Storage']:
    whichMemory(i)
    
# memory_df
df.info()
```
